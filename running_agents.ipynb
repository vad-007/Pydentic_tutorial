## Running Your Agents

Pydantic AI allows you to execute your agent in multiple ways depending on your use case.

Let's look at the three methods and when you should use them.

```python
import nest_asyncio
nest_asyncio.apply()

import asyncio
from pydantic_ai import Agent
from dotenv import load_dotenv

load_dotenv()
```

## 1.run_sync (Synchronous)

Best for: Simple scripts, data extraction, or when you just need the program to wait until the AI is finished thinking.

Let's use a summarizer agent.

```python
summarizer_agent = Agent(
    'groq:llama-3.1-8b-instant',
    system_prompt="Summarize the given text into exactly 3 words."
)

# The code stops here and waits for the AI.
summary_response = summarizer_agent.run_sync(
    "The quick brown fox jumps over the lazy dog in the green meadow under the bright sun."
)

print("Sync Output:", summary_response.output)  
```

## 2. run (Asynchronous)

Best for: Web backends (like FastAPI), or when you want your application to do other things while waiting for the AI.

Let's use a translation agent.

```python
translator_agent = Agent(
    'groq:llama-3.1-8b-instant',
    system_prompt="Translate the given text to Spanish."
)

translation_response = translator_agent.run_sync("Hello, how are you doing today?")
print("Sync Translation Output:", translation_response.output)

```

```python
async def get_async_translation():
    print("Sending request to AI...")
    # Your computer can do other things while 'awaiting' the answer!
    response = await translator_agent.run("Hello, how are you doing today?")
    print("Async Translation Output:", response.output)

# Run our async function
await get_async_translation ()
```
